# for all the pages

```python
all_info = {
    "item_name":[],
    "price":[],
    "rated_count":[],
    "old_price":[],
    "rated":[],
    "discount %":[]
}

page=1 

while page<=5: 
        url= f"https://www.jumia.com.ng/catalog/?q=watch&page={page}#catalog-listing"
        html=requests.get(url)
        soup=bs(html.content)


        containers = soup.findAll("div", {"class":"info"})
        for container in containers:
            try:
                item_name=container.find("h3", {"class" : "name"}).text
                all_info["item_name"].append(item_name)

            except:
                    all_info["item_name"].append(np.nan)

            try: 
                 price = container.find("div", {"class" : "prc"}).text
                 all_info["price"].append(price)

            except:
                all_info["price"].append(np.nan)

            try:
                old_price= container.find("div", {"class" : "old"}).text
                all_info["old_price"].append(old_price)

            except:
                all_info["old_price"].append(np.nan)

            try:
                rated_count = container.find("div", {"class": "rev"}).text.split("(")[1].strip(")")
                all_info["rated_count"].append(rated_count)

            except:
                all_info["rated_count"].append(np.nan)


            try:
                rated= container.find("div", {"class": "rev"}).text.split("(")[0].split()[0]
                all_info["rated"].append(rated)

            except:
                all_info["rated"].append(np.nan)

            try:
                discount_cent= container.find("div", {"class":"bdg _dsct _sm"}).text
                all_info["discount %"].append(discount_cent)
            except:
                all_info["discount %"].append(np.nan)

        
        page +=1     
df=pd.DataFrame(all_info)
df
```

```python
df.to_csv("jumia_watch.csv",index=False)
```
