```python
container=soup.findAll("div",  {"class":"a-section a-spacing-base"})
```

```python
soup.find("div",  {"class":"a-section a-spacing-base"})
```

```python
name =[]
price = []
for i in container:
    try:
        name.append(i.find("span", class_"a-size-base-plus a-color-base a-text-normal"),text)
    except:
        name.append("no name")
        
    try:
        price.append(i.find("span, class_="a-price_whole").text)
    except:
        price.append("no price")
                        
```

```python
import pandas as pd
d = {"watch name":name, "watch price: price"}
y= pd.dataframe(d)
```
