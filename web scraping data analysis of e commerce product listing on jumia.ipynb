# putting all item together for a single page

```python
all_info = {
    "name":[],
    "price":[],
    "old_price":[],
    "rating":[],
    "shipped":[]
}
containers=soup.findAll("div",  {"class":"info"})
for container in containers:
    try:
        name = container.find("h3", {"class":"name"}).text
        all_info["name"].append(name)
        
    except:
        all_info["name"].append(np.nan)
        
    try:
        price = container.find("div", {"class" : "prc"}).text
        all_info["price"].append(price)
        
    except:
        all_info["price"].append(np.nan)
        
    try:
        old_price=container.find("div", {"class" : "old"}).text
        all_info["old_price"].append(old_price)
        
    except:
        all_info["old_price"].append(np.nan)
            
        
    try:
        rating = container.find("div", {"class" :"bdg _dsct _sm"}).text
        all_info["rating"].append(rating)
        
    except:
        all_info["rating"].append(np.nan)
        
    try:
        shipped = container.find("div", {"class": "bdg _glb _xs"}).text
        all_info["shipped"].append(shipped)
        
    except:
        all_info["shipped"].append(np.nan)
```

```python
pd.DataFrame(all_info)
```

# finding all pages

```python
all_info = {
    "name":[],
    "price":[],
    "old_price":[],
    "rating":[],
    "shipped":[]
}

page=1 

while page<=5: 
            url= f"https://www.jumia.com.ng/catalog/?q=bag&page={page}#catalog-listing"
            html=requests.get(url)
            soup=bs(html.content)

            containers=soup.findAll("div",  {"class":"info"})
            for container in containers:
                try:
                    name = container.find("h3", {"class":"name"}).text
                    all_info["name"].append(name)

                except:
                      all_info["name"].append(np.nan)

                try:
                    price = container.find("div", {"class" : "prc"}).text
                    all_info["price"].append(price)

                except:
                     all_info["price"].append(np.nan)

                try:
                    old_price=container.find("div", {"class" : "old"}).text
                    all_info["old_price"].append(old_price)

                except:
                      all_info["old_price"].append(np.nan)


                try:
                    rating = container.find("div", {"class" :"bdg _dsct _sm"}).text
                    all_info["rating"].append(rating)

                except:
                      all_info["rating"].append(np.nan)

                try:
                    shipped = container.find("div", {"class": "bdg _glb _xs"}).text
                    all_info["shipped"].append(shipped)

                except:
                     all_info["shipped"].append(np.nan)




            page +=1     
df=pd.DataFrame(all_info)
df 
```

```python
df.to_csv("jumia_bag.csv",index=False)
```
